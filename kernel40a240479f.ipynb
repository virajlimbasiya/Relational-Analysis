```python
train = pd.read_table(r'/kaggle/input/kbp37-dataset/train.txt',header = None)
train.columns = ['sentence','relation']
```

```python
train.shape
```

```python
i=0;
while(i<train.shape[0]):
    train['sentence'][i] = train['relation'][i]
    train['relation'][i] = train['sentence'][i+1]
    i = i+2
train = train.iloc[::2]
train.index = train.index/2
```

```python
train.head()
```

```python
train['sentence'][1].split()
first = []
second = []
flag=0
flag1=0
for i in range(train.shape[0]):
    temp1 = []
    temp = []
    for x in train['sentence'][i].split():
        if flag==1 and x!='</e1>':
            temp.append(x)
        if flag1==1 and x!='</e2>':
            temp1.append(x)
        if x=='<e1>':
            flag=1
        elif x=='</e1>':
            flag=0
        elif x=='<e2>':
            flag1=1
        elif x=='</e2>':
            flag1=0
    train['relation'][i] = train['relation'][i][:len(train['relation'][i])-7]
    first.append([''.join(i) for i in temp])
    second.append([''.join(i) for i in temp1])
```
